```python
import uuid
import random
total_data = []
cur_group_id = "GroupID_9106a6c0-7ed2-488e-83db-149993c458ba"
group_mems = employee_data[employee_data["group_id"] == cur_group_id]
for i in range(2):
        rand_idx = random.randint(0, len(group_mems) - 1)
        name = group_mems.iloc[rand_idx]["hovaten"]
        emp_id = group_mems.iloc[rand_idx]["maso_doanvien"]
        # gender = group_mems.iloc[rand_idx]["gioitinh"]
        # gender = "Nam" if gender == "nam" else "Nữ"
        list_choices = ["Cháu lên ba", "Con cò bé bé", "Cháu yêu bà", "Ba thương con", "Chị ong nâu"]
        song = random.choice(list_choices)
        expect_time = random.choice([str(i) + " phút" for i in range(2, 10)])
        key = str(uuid.uuid4())
        cur_data = {
                "key": key,
                "Mã nhân viên": emp_id,
                "Tên": name,
                "Tiết mục": song,
                "Thời gian dự tính": expect_time,
        }
        total_data.append(cur_data)
```

```python
import json
json_total_data = json.dumps(total_data, ensure_ascii=False).replace("'", '"')
with open("test.json", "w", encoding="utf-8") as f:
        f.write(json_total_data)
```
